el tipo de sentencia mas conocido es el if

```python
x = int(input("ingresa un numero:"))
if x % 2 == 0:
   print("el numero es par.")
else:
    print("el numero es impar.")
```

```python
x = int(input("ingresa un numero:"))
if x % 3 == 0:
   print("el numero es multiplo de 3")
elif x % 3 == 1:
    print("no es multiplo de 3 porque el residuo es 1")
else:
    print("no es multiplo de 3 porque el residuo es 2")
```
